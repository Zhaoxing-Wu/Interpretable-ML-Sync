```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.context(['seaborn-paper'])
plt.rcParams.update({'font.size': 12})
```

# FCA - Caltech, UCLA, NWS

```python
fca_caltech[['# Dictionaries', 'Acc10', 'Acc20', 'Acc30']]
```

```python
fca_caltech = pd.read_csv('fca_Caltech36.csv')
fca_caltech = fca_caltech.set_index([pd.Index([2, 4, 8, 16])])

ax = fca_caltech.plot.line(marker='*')
plt.xlabel('Number of Dictionaries')
plt.ylabel('Accuracy')
plt.xscale('log', base=2)
plt.xticks([2, 4, 8, 16])
ax.get_xaxis().set_major_formatter(
    mpl.ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.yaxis.set_major_formatter(
    mpl.ticker.FormatStrFormatter('%.2f'))
ax.legend(title='Accuracy', labels=['k=10', 'k=15', 'k=20', 'k=25', 'k=30'],
          loc='upper center', bbox_to_anchor=(0.5, 1.25),
          ncol=3, fancybox=True, shadow=True)
plt.title("FCA - Caltech36", pad=55)
plt.show()
```
